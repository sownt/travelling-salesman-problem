```python
def visualize(matrix, label=True):
    G = nx.from_numpy_matrix(matrix, create_using=nx.DiGraph)
    layout = nx.spring_layout(G)
    nx.draw(G, layout, with_labels = True)
    if label:
        nx.draw_networkx_edge_labels(G, pos=layout)
    plt.show()
```

```python
case1 = np.array([[0,25,25,13],
               [25,0,25,13],
               [25,25,0,13],
               [13,13,13,0]])
```

```python
visualize(case1)
```

```python
def nn(G):
    V = len(G)
    visited = [False] * V
    visited[0] = True
    
    path = [0,]
    cost = 0
    for i in range(V):
        print(visited, i)
        minw = 0
        mine = -1
        for j in range(V):
            if visited[j] == False:
                minw = G[i][j]
                mine = j
        if mine != -1:
            path.append(mine)
            cost += minw
            visited[mine] = True
    
    return path, cost
```
